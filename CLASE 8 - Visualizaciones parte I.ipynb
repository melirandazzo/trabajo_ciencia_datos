```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

fig,ax = plt.subplots(figsize =(8,7))
fig.set_facecolor('#b2b2b2') #cambia el fondo del grfico, por defecto es blanco, se puede usar directamente grey
# leer ms en https://taniaizquierdo.com/color-hexadecimal/#:~:text=El%20color%20hexadecimal%20es%20un,por%20sus%20siglas%20en%20ingl%C3%A9s).

"""
#b2b2b2 es un valor hexadecimal que representa un color en el modelo de colores RGB (rojo, verde y azul)
utilizado en la programacin y el diseo. Los valores hexadecimales se utilizan para representar los valores de los componentes rojo, verde y azul
(en ese orden) en un solo nmero de 6 dgitos.

En este caso, #b2b2b2 representa un gris medio, donde el componente rojo tiene un valor de b2 en hexadecimal (178 en decimal),
el componente verde tiene un valor de b2 en hexadecimal (178 en decimal) y el componente azul tiene un valor de b2 en hexadecimal (178 en decimal).

Este valor hexadecimal se puede utilizar en diferentes contextos en los que se necesite especificar un color,
como en el fondo de una figura o en el color de una lnea o una barra en un grfico.

"""

ax.pie(data, labels = cars,radius=0.7, startangle=90, explode= (0,0,0,0,0,0.1), autopct= "%1.1f%%", colors=sns.color_palette("hls",7))

"""
labels: lista con los nombres de las etiquetas que se mostrarn en el grfico.
radius: radio del grfico circular.
startangle: ngulo inicial de la primera divisin del grfico.
explode: lista que especifica la separacin de cada divisin del grfico circular respecto al centro.
autopct="%1.1f%%" significa que se mostrar el valor porcentual con un decimal y un signo de porcentaje
colors: es una variable que se utiliza para almacenar una paleta de colores generada con la funcin color_palette() del paquete seaborn en Python.
"""
plt.legend(loc="lower right",labels=cars,fontsize=10)
plt.show()
```

# Grafico 1: histograma para ver distribucion del salario

```python
# Orientado a objetos
mpl.style.use('classic')
"""
establece el estilo de visualizacin predeterminado de Matplotlib como 'classic'.
 Matplotlib como vimos en clase tiene varios estilos de visualizacin incorporados, como 'ggplot', 'bmh', etc.
"""
fig, ax = plt.subplots(figsize=(10,6))
"""
Esta lnea crea una figura y un conjunto de ejes en los que se trazar el histograma.
La figura es el lienzo en el que se dibuja la visualizacin, mientras que los ejes son los objetos que se utilizan para
trazar los datos en la figura. Se establece el tamao de la figura en 10 pulgadas de ancho y 6 pulgadas de alto
"""
ax.hist(x= data.Salary, bins=None, orientation='vertical')
"""
Esta lnea traza el histograma en los ejes creados anteriormente.
La funcin hist() recibe varios argumentos, como los datos que se van a trazar
(en este caso, los salarios del conjunto de datos 'data'), el nmero de bins
(si no se especifica, Matplotlib calcular automticamente el nmero ptimo de bins),
y la orientacin del histograma ('vertical' o 'horizontal').
"""
ax.set_xlabel('Salario')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de Salario')

"""
Estas lneas hacen referencia a las etiquetas
"""
```

```python
# Orientado a estados
mpl.style.use('classic')
plt.figure(figsize=(10,6))
plt.hist(x= data.Salary, bins=None, orientation='vertical')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.title('Histograma de Salario')
```

Se logra ver que la distribucion de los datos es asimetrica a derecha con la mayoria de valores en la parte inferior de la distribucion (valores tipicos entre 50000-100000). Se logra ver tambien que hay presencia de valores por encima de 2000000
