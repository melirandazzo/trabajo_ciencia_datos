## Series de tiempo

> Tipo de dato Timestamp

```python
# Fecha clase de hoy
fecha = pd.to_datetime(pd.Timestamp.now())
fecha
```

```python
# Fecha próxima clase
fin = pd.to_datetime(fecha+ pd.Timedelta(weeks=1))
fin
```

```python
# Rango de fechas
fechas_1 = pd.date_range(start=fecha, end=fin,periods=8)
fechas_1
```

```python
mes_inicio = fecha.strftime('%Y/%m')
mes_inicio
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo con un índice de tipo fecha que incluye día, mes y año
df = pd.DataFrame(data=[1, 2, 3], index=pd.to_datetime(['01-01-2022', '02-01-2022', '01-03-2024'],dayfirst=True), columns=['Valor'])
# es clave usar dayfirst=True en este ejemplo
# Extraer el mes del índice
df['Mes'] = df.index.month
df['Año'] = df.index.year
df['Día'] = df.index.day

print(df)
```
