Utilice la desviación estándar para determinar qué tan dispersos están los datos con respecto a la media. Un valor de desviación estándar más alto indica una mayor dispersión de los datos.

```python
##caculando la desviación estandar.

df['age'].std().round(1)
```

```python
#Se contrasta std con el promedio, para saber si es un std alto o bajo.
df['age'].mean().round(1)
```

Es importante tener en cuenta que la desviación estándar depende de la escala de los datos. En este caso, si la escala de los datos es grande (por ejemplo, si se están midiendo valores en millones), una desviación estándar de 250 podría ser relativamente pequeña en comparación con los valores reales. Por otro lado, si la escala de los datos es pequeña (por ejemplo, si se están midiendo valores en centímetros), una desviación estándar de 250 podría ser relativamente grande en comparación con los valores reales.

```python
# Calcular el coeficiente de variación para cada columna
coef_var = df["age"].std() / df['age'].mean() *100
coef_var
```

No obstante, **no todas** las variables que tienen números son Cuantitativas, por lo que es necesario tener cuidado y considerar el contexto de los datos

```python
# cuando se aplica sobre variables cualitativas no puede calculas las
# metricas tradicionales, por lo que calcula la cuenta de valores únicos (unique),
# la moda, el valor más frecuente (top)
df.describe(include='category')
```

```python
# Para las variables categoricas es muy común realizar un conteo por categoria
# el método value_counts() permite contar las frecuencias de cada categoria
df.cut.value_counts()
```

```python
df.cut.value_counts(normalize=True) # En este caso calcula la proporción sobre el total de observaciones
```

```python
df.cut.value_counts(normalize=True)*100 # si lo multiplican por 100 esta en porcentaje
```

```python
# Vamos a cambiar el tipo de esta variable usando el método astype
df['cylinders'] = df['cylinders'].astype('category')
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.describe(include='category')
```

```python
# el método value_counts() permite contar las frecuencias de cada categoria
df['cylinders'].value_counts()
```

```python
# si utilizamos el parametro normalize se obtienen los datos
# con frecuencia relativa o porcentaje
df['cylinders'].value_counts(normalize=True)
```

```python
# el ID de los autos esta almacenado en el index, lo vamos a extraer y renombrar para poder comprobar su cardinalidad
df.reset_index(inplace=True)
df.rename(columns={'index':'ID'}, inplace=True)

# Para evaluar la cardinalidad de las variables podemos utilizar nunique()
# Este método cuenta los valores únicos de una variable
df.nunique()
```

Cuando la cardinalidad de una variable sea igual a la cantidad de filas, estamos tratando con un ID. <br>

```python
# Tambien se puede utilizar en una columna única
df.origin.nunique()
```

```python
# y tl método unique() muestra los datos únicos
df.origin.unique()
```

medidas de resumen

```python
promedio = df.acceleration.mean() # promedio o media
mediana = df.acceleration.median() # mediana o Q2
moda = df.acceleration.mode() # moda
primer_cuartil = df.acceleration.quantile(0.25) # Q1 cuartíl 1 --> percentíl 25
tercer_cuartil = df.acceleration.quantile(0.75) # Q3 cuartíl 2 --> percentíl 75
percentil_99 = df.acceleration.quantile(0.99) # percentíl 99


# Imprimir los resultados
print("Promedio: ", promedio)
print("Mediana: ", mediana)
print("Moda: ", moda)
print("Primer cuartil: ", primer_cuartil)
print("Tercer cuartil: ", tercer_cuartil)
print("Percentil 99: ", percentil_99)
```
