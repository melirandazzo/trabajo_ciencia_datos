# Diagrama de dispersion

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.scatterplot(x=data.Close, y= data.Open, marker='p')
plt.title('Relacion entre Precio apertura y cierre')
plt.xlabel('Precio de Cierre')
plt.ylabel('Precio de apertura')
```

```python
#Veamos una forma elegante de hacer una matriz de correlacin


# Seleccionar las dos columnas que quieres comparar
columnas_seleccionadas = data[['Close', 'Open']]

# Calcular la matriz de correlacin
matriz_correlacion = columnas_seleccionadas.corr()

# Crear una visualizacin de la matriz de correlacin utilizando seaborn
sns.heatmap(matriz_correlacion, cmap='coolwarm', annot=True)

"""
annot=True: este parmetro establece si deseas mostrar los valores exactos
en la matriz de correlacin. Si se establece en True, se mostrarn los valores en cada celda

cmap= es para seleccionar los colores dentro de la matriz, revisar opciones en https://seaborn.pydata.org/tutorial/color_palettes.html
"""

# Mostrar la visualizacin
plt.show()
```


Un valor de correlacin de 0.65 indica que hay una relacin moderada entre las dos variables.
Esto significa que cuando una variable aumenta, es probable que la otra variable tambin aumente,
y cuando una variable disminuye, es probable que la otra variable tambin disminuya.
Sin embargo, es importante tener en cuenta que la correlacin no implica causalidad,
lo que significa que no se puede inferir que una variable causa la otra solo porque estn correlacionadas.

Es importante tener en cuenta que el contexto de los datos y el propsito del anlisis pueden afectar la
interpretacin de la correlacin. En algunos casos, una correlacin de 0.65 puede considerarse alta y
en otros casos puede considerarse moderada.
Por lo tanto, es importante tomar en cuenta el contexto y el propsito del anlisis para interpretar adecuadamente
el valor de correlacin.
