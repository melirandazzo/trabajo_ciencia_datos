# Regresion lineal simple

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Ejemplo de datos: altura y peso
data = {'Altura': [150, 160, 170, 180, 190],
        'Peso': [50, 55, 60, 70, 80]}
df = pd.DataFrame(data)

# Variable dependiente (Y) y variable independiente (X)
X = df['Altura']# voy a utilizar la altura para hacerlo
Y = df['Peso']# lo que yo quiero predecir

# Agregar una constante para el t茅rmino de intersecci贸n en el modelo
X = sm.add_constant(X)

# Crear el modelo de regresi贸n
model = sm.OLS(Y, X).fit()

# Ver el resumen de la regresi贸n
print(model.summary())
```

### Interpretaci贸n del Modelo de Regresi贸n Lineal

1. **R-squared (R-cuadrado):**
   - **R-squared = 0.970**: El 97% de la variabilidad del peso puede explicarse por la altura. Esto indica que el modelo ajusta muy bien los datos.

2. **Adj. R-squared (R-cuadrado ajustado):**
   - **Adj. R-squared = 0.960**: Es similar al R-cuadrado, pero penaliza por variables innecesarias. Dado que solo tienes una variable, este valor es muy cercano a 1.

3. **F-statistic y Prob(F-statistic):**
   - **F-statistic = 96.43** y **Prob(F-statistic) = 0.00224**: Indican que el modelo es estad铆sticamente significativo, es decir, la altura afecta significativamente al peso.

4. **Coeficientes:**
   - **Constante (Intersecci贸n = intercepto (valor que toma Y cuando X vale 0)) = -64.5**: Este es el valor del peso cuando la altura es 0. En este contexto, no tiene mucho sentido pr谩ctico, pero sirve para ajustar la l铆nea.
   - **Altura = 0.75**: Por cada incremento de 1 cm en la altura, el peso aumenta en 0.75 kg.

5. **P-values (P>|t|):**
   - **Constante (0.016)** y **Altura (0.002)**: Ambos valores p son menores que 0.05, lo que indica que son significativos, es decir, relevantes para el modelo.

6. **Intervalos de confianza (95%):**
   - **Constante**: Va de -105.964 a -23.036.
   - **Altura**: Va de 0.507 a 0.993, lo que confirma que la relaci贸n entre altura y peso es significativa.

7. **Durbin-Watson:**
   - **Durbin-Watson = 1.429**: Un valor cercano a 2 sugiere que no hay autocorrelaci贸n significativa en los errores.

8. **Normalidad de los residuos:**
   - **Jarque-Bera (JB) = 0.375**, **Prob(JB) = 0.829**: No hay evidencia significativa de que los residuos no sigan una distribuci贸n normal.

### Conclusi贸n:
El modelo se ajusta bien a los datos. La altura tiene un impacto significativo en el peso, y no hay problemas importantes de autocorrelaci贸n o normalidad de los residuos.

# Regresion lineal multiple


```python
# Datos ficticios de ejemplo
data = {
    'Altura': [150, 160, 170, 180, 190],
    'Edad': [25, 35, 45, 55, 65],
    'Actividad_Fisica': [2, 3, 4, 2, 5],
    'Peso': [50, 60, 65, 75, 80]
}

df = pd.DataFrame(data)

# Variables independientes
X = df[['Altura', 'Edad', 'Actividad_Fisica']]

# Variable dependiente
Y = df['Peso']

# Agregar una constante (intercepto) al modelo
X = sm.add_constant(X)

# Ajustar el modelo de regresi贸n m煤ltiple
model = sm.OLS(Y, X).fit()

# Resumen del modelo
print(model.summary())
```

## Interpretaci贸n de los resultados de la regresi贸n lineal m煤ltiple

### Resumen del modelo:

- **Variable dependiente (y):** Peso
- **R-cuadrado (R-squared):** 0.992
    - El R-cuadrado indica que el modelo explica el 99.2% de la variabilidad de la variable dependiente (Peso), lo cual sugiere un ajuste excelente del modelo.
- **R-cuadrado ajustado:** 0.984
    - El R-cuadrado ajustado tiene en cuenta el n煤mero de predictores en el modelo. Con un valor de 0.984, sigue indicando un ajuste muy bueno.
- **F-statistic:** 121.5 (p-valor = 0.00816)
    - Este estad铆stico indica si el modelo en su totalidad es significativo. El p-valor menor que 0.05 indica que el modelo es estad铆sticamente significativo.
- **N煤mero de observaciones:** 5


### Coeficientes de los predictores:

- **Constante (Intercepto):** -0.0021 (p = 0.096)
    - La constante es negativa, pero no es significativa a nivel 0.05 ya que el p-valor es mayor que 0.05.

- **Altura:** 0.2642 (p = 0.005)
    - Un aumento en la Altura en 1 unidad se asocia con un aumento en el Peso de 0.2642 unidades. Este coeficiente es estad铆sticamente significativo, con un p-valor menor que 0.05.

- **Edad:** 0.5265 (p = 0.019)
    - Un aumento en la Edad en 1 unidad se asocia con un aumento en el Peso de 0.5265 unidades. Este coeficiente tambi茅n es significativo, ya que su p-valor es menor que 0.05.

- **Actividad F铆sica:** -0.8140 (p = 0.384)
    - Aunque el coeficiente es negativo, lo que sugerir铆a una relaci贸n inversa entre la actividad f铆sica y el peso, este predictor no es estad铆sticamente significativo (p-valor mayor que 0.05).

### Diagn贸sticos:

- **Durbin-Watson:** 2.852
    - Este valor se encuentra cerca de 2, lo que indica que no hay problemas serios de autocorrelaci贸n en los residuos.
- **Condici贸n del n煤mero (Cond. No.):** 1.34e+18
    - Este valor es muy alto, lo que sugiere posibles problemas de multicolinealidad entre las variables independientes.

### Conclusi贸n:
Este modelo tiene un muy buen ajuste con un R-cuadrado alto y significatividad estad铆stica general (p-valor del F-statistic < 0.05). Sin embargo, no todos los predictores son significativos (por ejemplo, Actividad F铆sica), y podr铆a haber problemas de multicolinealidad, lo que requiere mayor revisi贸n o modificaci贸n del modelo.

```python
traindataset = data[data["SalePrice"].notnull()]

# train and test split
X = traindataset.drop(columns=["SalePrice"], axis=1)
y = traindataset["SalePrice"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# 80 train y 20% test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Building Pipeline
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'SVR': SVR(),
    'RandomForestRegressor': RandomForestRegressor(),
    'XGBRegressor': XGBRegressor(objective='reg:squarederror'),
    'LGBMRegressor': LGBMRegressor(verbose=-1),
}

params = {
    'Ridge': {'model__alpha': [0.1, 1.0, 10.0]},
    'Lasso': {'model__alpha': [0.001, 0.01, 0.1]},
    'ElasticNet': {'model__alpha': [0.001, 0.01, 0.1],
                   'model__l1_ratio': [0.2, 0.5, 0.8]},
    'SVR': {'model__C': [0.1, 1, 10], 'model__epsilon': [0.1, 0.2, 0.5],
            'model__kernel': ['linear', 'rbf']},
    'RandomForestRegressor': {'model__n_estimators': [10, 50, 100]},
    'XGBRegressor': {'model__learning_rate': [0.1, 0.5],
                     'model__max_depth': [3, 5, 7]},
    'LGBMRegressor': {'model__learning_rate': [0.1, 0.5],
                      'model__num_leaves': [31, 50, 100]},
    'CatBoostRegressor': {'model__learning_rate': [0.01, 0.1, 0.5],
                          'model__depth': [4, 6, 8]}
}

results = []
```

```python
# Define the RMSE function
def rmse(y_true, y_pred):
    return math.sqrt(mean_squared_error(y_true, y_pred))

rmse_scorer = make_scorer(rmse, greater_is_better=False)
```

```python
for name, model in models.items():
    pipeline = Pipeline([
        ('model', model)
    ])
    if name in params:
        grid_search = GridSearchCV(pipeline, params[name], cv=5, scoring=rmse_scorer)
    else:
        grid_search = GridSearchCV(pipeline, {}, cv=5, scoring=rmse_scorer)

    grid_search.fit(X_train, y_train)
    y_pred = grid_search.predict(X_test)
    r2 = grid_search.best_estimator_.score(X_test, y_test)
    rmse_val = rmse(y_test, y_pred)
    results.append({
        'Model': name,
        'R-Squared Score': r2,
        'RMSE Score': rmse_val,
        'Best Parameters': grid_search.best_params_
    })

results_df = pd.DataFrame(results)
```
